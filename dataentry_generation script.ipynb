```python
market_code = "HI"
region_code = "MDXHAWAII"
df = pd.read_csv ('himdx.csv', header=None)
```

```python
month_strings = ["202201", "202202","202203", "202204", "202205", "202206" , "202207", "202208", "202209", "202210", 
                 "202211","202212"]
authtype_strings = ["Inpatient", "Inpatient Psych", "Sub-Acute", "Emergency Room", "Observation", "Outpatient Hospital", 
                    "Outpatient Dialysis", "Outpatient Surgery", "Ambulatory Surgical Center", "Ambulance", "DME & Prosthetics/Orthotics",
                    "Home Health", "Other Institutional", "Urgent Care", "Primary Care", "Specialist", "Ancillary/Other", "Pharmacy"]
```

```python
# df.loc[row_indexer,column_indexer]
with open('HI_MDXHAWAII.csv', 'w', newline='') as csvfile:
    spamwriter = csv.writer(csvfile, dialect='excel')
    for j in range(0, len(authtype_strings)):
        for k in range(0,len(month_strings)):
            row = [market_code, region_code, authtype_strings[j], month_strings[k], str(df.loc[j, k])]
            spamwriter.writerow(row)
```
